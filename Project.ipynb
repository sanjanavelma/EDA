```python
df['artwork_id'] = df['artwork_id'].str.replace(r'[A-Za-z]', '', regex=True)
df.head()
```

```python
df['creation_year'] = df['creation_year'].astype('Int64')
```

```python
df = df.rename(columns={'artwork_id_numeric': 'artwork_id'})
df.to_csv('your_dataset.csv', index=False)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
numeric_columns = df.select_dtypes(include=['number']).columns.tolist()
print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numeric_columns)
df.head() 
```

```python
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
```

```python
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
df['data_as_of'] = pd.to_datetime(df['data_as_of'], errors='coerce')
df['data_loaded_at'] = pd.to_datetime(df['data_loaded_at'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df.head()
```

```python
df.isnull().sum()*100/len(df)
```

```python
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
```

```python
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.isnull().sum()
```

```python
df['zip_code'] = df['zip_code'].astype(str)
```

```python
df['artwork_id'] = df['artwork_id'].astype(str)
```

```python
df.dtypes
```

```python
# Save cleaned data to a new CSV
df.to_csv('Art_collection', index=False)

print("Data cleaning complete!")
df.head(7)
```

# VISUALISATION

## Histogram of ZIP Codes
### Visualization Type: Histogram
### Purpose: To show the distribution of ZIP codes within the dataset.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df['zip_code'], bins=10, color='teal', edgecolor='black')
plt.title('Histogram of ZIP Codes')
plt.xlabel('ZIP Code')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

## Top 10 Art Types
### Visualization Type: Bar Chart
### Purpose: To show the most common art types in the dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
top_art_types = df['medium'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_art_types.index, y=top_art_types.values)
plt.title('Top 10 Art Types')
plt.xlabel('Art Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Distribution of Artwork by Artist
### Visualization Type: Bar Chart
### Purpose: To show the number of artworks created by each artist.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,30))
sns.countplot(data=df, y='artist', order=df['artist'].value_counts().index)
plt.title('Number of Artworks by Artist')
plt.xlabel('Number of Artworks')
plt.ylabel('Artist')
plt.show()
```

## Medium Used in Artworks
### Visualization Type: Pie Chart or Bar Chart
### Purpose: To visualize the proportion of different mediums used.

```python
plt.figure(figsize=(25,35))
df['medium'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Mediums Used in Artworks')
plt.ylabel('')
plt.show()
```

## Geographical Distribution of Artworks
### Visualization Type: Scatter Plot or Map Plot
### Purpose: To visualize the geographical distribution of artworks using latitude and longitude.

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['longitude'], df['latitude'], alpha=0.5)
plt.title('Geographical Distribution of Artworks')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## Creation Year Trend
### Visualization Type: Line Chart
### Purpose: To observe trends in the number of artworks created over the years.

```python
plt.figure(figsize=(12, 6))
df['creation_year'].value_counts().sort_index().plot.line(marker='o')
plt.title('Trend of Artworks Created Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Artworks')
plt.grid()
plt.show()
```

## Correlation Matrix
### Visualization Type: Heatmap
### Purpose: To understand the correlation between numerical features.

```python
plt.figure(figsize=(10, 8))
correlation = df.corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Artworks by Location
### Visualization Type: Bar Chart
### Purpose: To show the number of artworks based on the location or facility.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='facility', order=df['facility'].value_counts().index)
plt.title('Number of Artworks by Facility')
plt.xlabel('Number of Artworks')
plt.ylabel('Facility')
plt.show()
```
